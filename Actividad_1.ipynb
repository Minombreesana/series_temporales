1. Generar serie de tiempo y graficarla

```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mean = 0
std = 1 
num_samples = 500
X = numpy.random.normal(mean, std, size=num_samples)

plt.plot(X)
plt.show()
```

```python
df = pd.DataFrame(X)
df.describe()
```

```python
plt.hist(X, bins = 30)
plt.show()
```
